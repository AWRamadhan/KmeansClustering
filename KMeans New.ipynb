```python
def splitData(x,y):
    val1 = np.random.randint(0, len(x))
    val2 = np.random.randint(val1, len(x))
    while val2 - val1 != len(x)*0.2:
        val1 = np.random.randint(0, len(x))
        val2 = np.random.randint(val1, len(x))
    return val1, val2
```

```python
X_all = []
Y_all = []
for idx in X_coord:
    X_all.append(idx)
    
for idx in Y_coord:
    Y_all.append(idx)

actual = []
for idx in actual_val:
    actual.append(idx)
    
    
test_begin, test_stop = splitData(X_all,Y_all)

X = []
Y = []
for i in range(test_begin):
    X.append(X_all[i])
for i in range(test_begin):
    Y.append(Y_all[i])

for test_stop in range(len(X_all)):
    X.append(X_all[test_stop])
for test_stop in range(len(Y_all)):
    Y.append(Y_all[test_stop])

X_test = []
Y_test = []

for test_begin in range(test_stop):
    X_test.append(X_all[test_begin])
for test_begin in range(test_stop):
    Y_test.append(Y_all[test_begin])

actual_value = []
for test_begin in range(test_stop):
    actual_value.append(actual[test_begin])
```

```python
import random
def getRandomPoints(total_centroid):
    c_points = []
    randomX = sorted(set(X))
    searchX = []
    for i in range(len(randomX)):
        if len(randomX)%50:
            searchX.append(randomX[i])
    for point in range(total_centroid):     
        point_hold = []
        xpoint = random.choice(searchX)
        ypoint = random.choice(Y)
        point_hold.append(xpoint)
        point_hold.append(ypoint)
        c_points.append(point_hold)
    
    return c_points
```

```python
import math
def getEucludianDistance(xpoints_data,xpoints_centroid,ypoints_data,ypoints_centroid):
    return math.sqrt((xpoints_data-xpoints_centroid)**2+(ypoints_data-ypoints_centroid)**2)

def calculateEucludianCentroid(centroid, X, Y):
    centroid_points = []
    c_val = []
    for urutan in range(len(X)):
        calculate = getEucludianDistance(X[urutan],centroid[0],Y[urutan],centroid[1])
        if calculate == 0 :
            c_val.append(1)
        else :
            c_val.append(calculate)
    return c_val
```

```python
def compareCentroid(c_points ,X):
    for first in range(0, len(c_points)):
        if first != len(c_points)-1:
            for second in range(first+1, len(c_points)):
                for count in range(len(X)):
                    if (c_points[first][count] > c_points[second][count]):
                        c_points[second][count] = 0
                    else:
                        c_points[first][count] = 0
    return c_points
```

```python
def makeItCenter(list):
    coord = []
    for c_count in range(len(list)):
        count = 0
        xpoint = 0
        ypoint = 0
        for data_count in range(len(X)):
            if (list[c_count][data_count] != 0):
                xpoint += X[data_count]
                ypoint += Y[data_count]
                count += 1
        if count==0 :
            count = 1
        coord.append(xpoint/count)
        coord.append(ypoint/count)
    
    return coord
```

```python
def cluster(centroid_data, X):
    position = []
    color = 0
    for c_count in range(len(centroid_data)):
        coord = []
        color += 1
        for data_count in range(len(X)):
            c_postion = []
            if (centroid_data[c_count][data_count] != 0):
                c_postion.append(X[data_count])
                c_postion.append(Y[data_count])
                c_postion.append(color)
                coord.append(c_postion)
        position.append(coord)
        
    return position
```

```python
def KMeansLoop(n):
    centroid = getRandomPoints(n)
    random_point = []
    check = True
    loop = 0
    while check:
        old_point = random_point
        calculate = []

        for j in range(len(centroid)):
            calculate.append(calculateEucludianCentroid(centroid[j], X, Y))

        compare = compareCentroid(calculate ,X)            
        random_point = makeItCenter(compare)
        centroid = [random_point[i * 2:(i + 1) * 2] for i in range((len(random_point) + 2 - 1) // 2 )]

        cluster_member = cluster(compare, X)

        if min(random_point) != 0 :
            check = False

        for i in range(len(cluster_member)):
            if len(cluster_member[i]) == 1:
                check = True

        if min(random_point)==0:
            centroid = getRandomPoints(n)
            check = True
        
    return random_point, cluster_member
```

```python
import statistics 
from collections import Counter
def median(l):
    if len(set(l)) == len(l): 
        return max(set(l), key=l.count)
    else :
        hold = 0
        counter = Counter(l)
        max_count = max(counter.values())
        mode = [k for k,v in counter.items() if v == max_count]
        for i in range(len(mode)):
            hold += mode[i]
        return hold/len(mode)
```

```python
def meanSquareError(centroid, cluster_member):
    data_test = []
    class_test = []
    for i in range(len(cluster_member)):
        hold = []
        if list(actual_val1)[0] == list(actual_val)[0]:
            for j in range(len(cluster_member[i])):
                hold.append(cluster_member[i][j][2])
        else :
            for j in range(len(cluster_member[i])):
                hold.append((cluster_member[i][j][0]+cluster_member[i][j][1])//2 )
        class_test.append(cluster_member[i][j][2])
        data_test.append(hold)
        
    predict = []
    for i in range(len(data_test)):
        predict.append(median(data_test[i]))

    data_pred = []
    calculate = []
    centroid = [centroid[i * 2:(i + 1) * 2] for i in range((len(centroid) + 2 - 1) // 2 )]
    for j in range(len(centroid)):
        calculate.append(calculateEucludianCentroid(centroid[j], X_test, Y_test))
    compare = compareCentroid(calculate ,X_test)            
    cluster_member = cluster(compare, X_test)
    if list(actual_val2)[0] == list(actual_val)[0]:
        for i in range(len(cluster_member)):
            for j in range(len(cluster_member[i])):
                data_pred.append((cluster_member[i][j][0]+cluster_member[i][j][1])//2)
    else:
        for i in range(len(cluster_member)):
            for j in range(len(cluster_member[i])):
                if cluster_member[i][j][2] == class_test[i]:
                    data_pred.append(predict[i])
                
    count = 0
    mean_square = 0
    for i in range(len(actual_value)):
        hold = (data_pred[i] - actual_value[i])**2
        mean_square = mean_square + hold
        if data_pred[i] == actual_value[i]:
            count += 1

    accuration = count/len(actual_value)
    mean_square_error = mean_square/len(actual_value)

    return accuration, mean_square_error
```

```python
data_error = []
n = [3,4,5,6]
for i in range(len(n)):
    hold = 0
    print('Running on K : ',n[i])
    for j in range(3):
        val1 ,val2 = KMeansLoop(n[i])
        hold += meanSquareError(val1,val2)[1]
    data_error.append(hold)
    print('------Succes------')

for i in range(len(n)):
    if min(data_error) == data_error[i]:
        bestK = i
```
